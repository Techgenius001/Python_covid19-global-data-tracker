## 1️⃣ Data Collection

```python
import pandas as pd

# Load the dataset directly from the internet
url = 'https://covid.ourworldindata.org/data/owid-covid-data.csv'
df = pd.read_csv(url)

# Preview the data
df.head()
```

## 2️⃣ Data Loading & Exploration

```python
# Check column names and data types
print("Columns in the dataset:\n", df.columns)
print("\nData types:\n", df.dtypes)

# Basic info about the dataset
print("\nDataset Info:")
df.info()

# Check for missing values
print("\nMissing values per column:\n", df.isnull().sum())

# Summary statistics for numeric columns
df.describe()
```

## 3️⃣ Data Cleaning

```python
# Filter for selected countries
countries = ['Kenya', 'United States', 'India']
df_filtered = df[df['location'].isin(countries)]

# Convert date column to datetime
df_filtered['date'] = pd.to_datetime(df_filtered['date'])

# Sort data by country and date
df_filtered = df_filtered.sort_values(['location', 'date'])

# Drop rows where total_cases or total_deaths is missing
df_filtered = df_filtered.dropna(subset=['total_cases', 'total_deaths'])

# Optionally: fill missing values in other columns with interpolation
df_filtered = df_filtered.interpolate()

# Confirm the cleaning
print("Cleaned Data Sample:\n", df_filtered.head())
```

## 4️⃣ Exploratory Data Analysis

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set Seaborn style
sns.set(style="darkgrid")

# Plot total cases over time
plt.figure(figsize=(12,6))
for country in countries:
    country_data = df_filtered[df_filtered['location'] == country]
    plt.plot(country_data['date'], country_data['total_cases'], label=country)

plt.title('Total COVID-19 Cases Over Time')
plt.xlabel('Date')
plt.ylabel('Total Cases')
plt.legend()
plt.tight_layout()
plt.show()

# Plot total deaths over time
plt.figure(figsize=(12,6))
for country in countries:
    country_data = df_filtered[df_filtered['location'] == country]
    plt.plot(country_data['date'], country_data['total_deaths'], label=country)

plt.title('Total COVID-19 Deaths Over Time')
plt.xlabel('Date')
plt.ylabel('Total Deaths')
plt.legend()
plt.tight_layout()
plt.show()

# Plot daily new cases
plt.figure(figsize=(12,6))
for country in countries:
    country_data = df_filtered[df_filtered['location'] == country]
    plt.plot(country_data['date'], country_data['new_cases'], label=country)

plt.title('Daily New COVID-19 Cases')
plt.xlabel('Date')
plt.ylabel('New Cases')
plt.legend()
plt.tight_layout()
plt.show()

# Calculate death rate and plot
df_filtered['death_rate'] = df_filtered['total_deaths'] / df_filtered['total_cases']

plt.figure(figsize=(12,6))
for country in countries:
    country_data = df_filtered[df_filtered['location'] == country]
    plt.plot(country_data['date'], country_data['death_rate'], label=country)

plt.title('COVID-19 Death Rate Over Time')
plt.xlabel('Date')
plt.ylabel('Death Rate')
plt.legend()
plt.tight_layout()
plt.show()
```

## 5️⃣ Vaccination Visualization

```python
# Plot cumulative vaccinations over time
plt.figure(figsize=(12,6))
for country in countries:
    country_data = df_filtered[df_filtered['location'] == country]
    plt.plot(country_data['date'], country_data['total_vaccinations'], label=country)

plt.title('Cumulative COVID-19 Vaccinations Over Time')
plt.xlabel('Date')
plt.ylabel('Total Vaccinations')
plt.legend()
plt.tight_layout()
plt.show()

# Calculate % of population vaccinated
# We'll use 'people_fully_vaccinated_per_hundred' which represents % of fully vaccinated people

plt.figure(figsize=(12,6))
for country in countries:
    country_data = df_filtered[df_filtered['location'] == country]
    plt.plot(country_data['date'], country_data['people_fully_vaccinated_per_hundred'], label=country)

plt.title('Fully Vaccinated People (% of Population)')
plt.xlabel('Date')
plt.ylabel('% Fully Vaccinated')
plt.legend()
plt.tight_layout()
plt.show()
```

## 7️⃣ Insights & Reporting

## 📊 Key Insights

1. **Vaccination Progress**: Among Kenya, India, and the USA, the USA has consistently maintained the highest vaccination rates. Kenya’s rollout is significantly slower but improving steadily.

2. **Cases and Deaths**: The USA recorded the highest total cases and deaths. However, India showed massive case spikes in mid-2021, while Kenya maintained relatively lower numbers.

3. **New Cases Trends**: Daily new cases show clear waves, aligning with major global variant outbreaks (e.g., Delta, Omicron).

4. **Death Rate**: Despite high total deaths, the USA’s death rate as a percentage of cases is slightly lower than that of Kenya, possibly indicating better healthcare capacity.

5. **Data Anomalies**: Some missing data points (especially in early 2020) suggest delays or inconsistencies in early data collection across countries.
